# Energy Consumption and CO2 Emission Prediction for Residential Buildings

## 1. Problem Statement:
CO2 Emission and Energy Consumption are one of the major factors affecting the environment. As a homeowner or a renter it is really hard to get an estimate of what an annual energy consumption of their home will be like. Also apart from that it would be good if the user will get any suggestions to reduce their annual energy consumption which will help to reduce the amount of bills they are paying. Also , in this time of climate change it is essential to know the CO2 Emission from your home/building which is a major contributing factor of the environemental issues.

This project will be the attempt to predict/estimate the annual energy(KW/H) and CO2 emission of a residential building using the models trained on real world dataset, which I hope will help the end users find it useful to get valuable insights of their own homes and how to improve their energy consumption

## 2. Dataset :
The dataset we are using for this project is called `Energy Performance Certificate(EPC)` dataset from the website of the UK's official Department for Levelling Up, Housing & Communities. This is an open data published by the UK Government and had published the data from 2008 Oct 1, and are updated monthly. The data can be downloaded from here: https://epc.opendatacommunities.org

### 2.1 Dataset Description
This is a very big dataset which contains the Energy Performance data of each residential buildings in the England and Wales, and we have the option to download the data of each towns from the England and Wales.
The data are of two types :
* Domestic EPC (Residential Buildings and Houses)
* Non-Domestic EPC (Non-Residential Buildings, Offices, Businesses etc.)

Since we are focusing on Residential Buildings we are using the Domestic EPCs .

We are using Sqlite to create a connection and to combine all of our 3 dataset from different cities (London, Birmingham and Manchester)

```python
import sqlite3
#Creating connection to sqlite database
connection = sqlite3.connect("Energy_Data.db")
cursor = connection.cursor()

#Writing each dataframe to SQLite Database as a table
df_london.to_sql("London",connection,if_exists='replace',index=False)
df_birmingham.to_sql("Birmingham",connection,if_exists='replace',index=False)
df_manchester.to_sql("Manchester",connection,if_exists='replace',index=False)

connection.commit()
```

### Combining the data using UNION ALL query and storing it as `energy_data`

```python
#Writing SQL Query to combine all three tables
query =  '''
        SELECT * FROM London
        UNION ALL
        SELECT * FROM Birmingham
        UNION ALL
        SELECT * FROM Manchester;
        '''
#Excecuting the query and storing the result in a pandas dataframe
energy_data = pd.read_sql_query(query,connection)

#Display the combined data
print(energy_data)
```

Functionising the cleaning of Construction Age Band

```python
import numpy as np
# Defining Age Bands
age_bands = {
    "Before 1900": range(0,1900),
    "1900-1929": range(1900,1930),
    "1930-1949": range(1930,1950),
    "1950-1966": range(1950,1967),
    "1967-1975": range(1967,1976),
    "1976-1982": range(1976,1983),
    "1983-1990": range(1983, 1991),
    "1991-2002": range(1991, 2003),
    "2003-2010": range(2003, 2011),
    "2011-2020": range(2011, 2021),
    "2021 onwards": range(2021, 2030)
}

def clean_construction_age(value):
    if pd.isna(value) or value in ['NO DATA!','INVALID!']:
          return np.nan #Filling missing values
    if "England and Wales" in str(value):
        return value.replace("England and Wales: ", "").strip() #Remove prefix
    if str(value).isdigit():
        year = int(value)
        for band,years in age_bands.items():
            if year in years:
                return band
    return value

#Using Cleaning function to clean the age band
df['CONSTRUCTION_AGE_BAND'] = df['CONSTRUCTION_AGE_BAND'].apply(clean_construction_age)

df['CONSTRUCTION_AGE_BAND'].fillna("Unknown", inplace=True)
```

```python
#Clean NO DATA! and similar value by replacing in with Nan
def clean_data(df):
    df = df.replace(['NO DATA!', 'NODATA!', 'INVALID!', 'Unknown', 'nan'],np.nan)
    return df
def convert_numerical_to_string(df,categorical_columns):
    for column in categorical_columns:
        df[column] = df[column].apply(lambda x:str(x) if isinstance(x,(float,int)) else x)
    return df      

def extract_datetime_features(df, datetime_columns):
    for column in datetime_columns:
        df[column] = pd.to_datetime(df[column], errors='coerce')  # Ensure valid datetime format
        df[column + '_year'] = df[column].dt.year
        df[column + '_month'] = df[column].dt.month
        df[column + '_day'] = df[column].dt.day
        df[column + '_weekday'] = df[column].dt.weekday
    return df
```

**N**ow we have trained our Model and Got some good results , Now we are entering into the phase of model deployement. As we know we have trained our initial model with 59 features , but when it comes to the production/deployement , we need to create a User-friendly web app which will give the user there property's annual energy consumption current. An end-user might not want to provide all the 59 features to get the prediction , so we need to focus on the basic details which the homeowners/renters are likely to know. For that we can divide the features into three categories:
- **General Information:**
    * PROPERTY_TYPE (House, Flat, Bungalow, etc.) 
    * BUILT_FORM (Detached, Semi-detached, Terraced, etc.)
    * TOTAL_FLOOR_AREA (Can be estimated or found in property documents)
    * NUMBER_HABITABLE_ROOMS(How many rooms are used for living/sleeping?)
    * NUMBER_HEATED_ROOMS(Rooms with heating systems)
- **Basic Energy Features:**
    * MAIN_FUEL (Gas, electricity, oil, etc.)
    * LOW_ENERGY_LIGHTING (Are LED bulbs used? Yes/No)
    * MAINHEAT_DESCRIPTION (Main heating type, e.g., boiler, electric heater)
    * MAINS_GAS_FLAG (Do they use gas for heating?)
    * SOLAR_WATER_HEATING_FLAG (Do they have solar water heating?)
- **Windows & Insulation:**
    * GLAZED_TYPE (Single-glazed, Double-glazed, etc.)
    * WALLS_DESCRIPTION (Cavity wall, solid wall, etc., might be known from property documents)
